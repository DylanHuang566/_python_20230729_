```python
#import matplotlib.pyplot as plt
#from matplotlib.font_manager import FontProperties
#font=FontProperties(fname='/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf',size=8)
#font1=FontProperties(fname='/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf',size=12)
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)

#student1 = {'':78, '':80, '':92, '':75, '':85}
#student2 = {'':82, '':90, '':88, '':89, '':91}

#subjects = list(student1.keys())
#value1 = list(student1.values())
#value2 = list(student2.values())

#figure = plt.figure(figsize=(8.0,5.0),dpi=250.0,facecolor="green")
#ax = figure.add_subplot()
#ax.plot(subjects, value1,)
#ax.plot(subjects, value2)
#plt.show()



#codespace 
#:sudo apt-get update -y
#:sudo apt-get install -y fonts-wqy-zenhei
#install -y
#:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
font1=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=12)
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)

student1 = {'':78, '':80, '':92, '':75, '':85}
student2 = {'':90, '':75, '':98, '':65, '':75}

subjects =list(student1.keys())
value1 =list(student1.values())
value2 = list(student2.values())

fig = plt.figure(figsize=(8,5), dpi=250, facecolor="#cccccc")
ax = fig.add_subplot(1,1,1)
ax.set_title("12",fontproperties=font)
ax.set_xlabel("",fontproperties=font)
ax.set_ylabel("",fontproperties=font)
ax.plot(subjects,value1,'m-.o')
ax.plot(subjects,value2,'c--o')
ax.set_xticklabels(subjects,fontproperties=font)
ax.legend(["1","2"],prop=font)

x_locations = ax.get_xticks()
#y_locations = ax.get_yticks()

for i in range(len(x_locations)):
    ax.annotate(str(value1[i]),(x_locations[i]+0.03,value1[i]))
    ax.annotate(str(value2[i]),(x_locations[i]+0.03,value2[i])) 
      
value1_total = 0
value2_total = 0
for score in value1:
    value1_total += score

for score in value2:
    value2_total += score

ax.annotate(f'1{value1_total}\n2{value2_total}',(2.5,85),fontproperties=font)
ax.legend([f"1:{value1_total}",f"2:{value2_total}"],prop=font)  
plt.show()
```
